```r
size1 <- function(n){
    size <- 0
    limit <- floor(2.5*n)
    if (limit == 2.5*n){
        limit = limit-1
    }
    for(i in 0:limit){
        #para não quebrar fatorial
        if(i>170){
            break
        }
        #para não quebrar a fdp poisson
        if( ((n^i)*exp(-n)/factorial(i)) == Inf){
            break
        }
        size <- size + ((n^i)*exp(-n)/factorial(i))
    }
    
    return (1-size)
}

size2 <- function(n){
    size <- 0
    limit <- floor(3*n) - 1
    for(i in 0:limit){
        #para não quebrar fatorial
        if(i>170){
            break
        }
        #para não quebrar poisson
        if( ((n^i)*exp(-n)/factorial(i)) == Inf){
            break
        }
        size <- size + ((n^i)*exp(-n)/factorial(i))
        
    }
    return (1-size)
}
```

```r
sizes_1 <- c()
sizes_2 <- c()

for(i in 1:100){
    sizes_1 <- c(sizes_1, size1(i))
    sizes_2 <- c(sizes_2, size2(i))
}

plot(1:100, sizes_1, type="l",lty=1, xlab="Tamanho amostral", ylab="Tamanho do teste")
lines(1:100, sizes_2,lty=2)
legend(80, 0.08, c(expression("k"[1]), expression("k"[2])), lty=c(1,2))
```

```r
power1 <- function(n){
    power <- 0
    limit <- floor(2.5*n)
    if(limit == 2.5*n){
        limit <- limit-1
    }
    for(i in 0:limit){
        if(i>170){
            break
        }
        if( ((3*n^i)*exp(-3*n)/factorial(i)) == Inf){
            break
        }
        power <- power + ((3*n^i)*exp(-3*n)/factorial(i))
    }
    
    return (1-power)
}

power2 <- function(n){
    power <- 0
    limit <- floor(3*n) - 1
    for(i in 0:limit){
        if(i>170){
            break
        }
        if( ((3*n^i)*exp(-3*n)/factorial(i)) == Inf){
            break
        }
        power <- power + ((3*n^i)*exp(-3*n)/factorial(i))
        
    }
    return (1-power)
}
```

```r
powers_1 <- c()
powers_2 <- c()

for(i in 1:100){
    powers_1 <- c(powers_1, power1(i))
    powers_2 <- c(powers_2, power2(i))
}

plot(1:100, powers_1, type="l",lty=1, xlab="tamanho amostral", ylab="poder do teste")
lines(1:100, powers_2, lty=2)
legend(80, 0.08, c(expression("k"[1]), expression("k"[2])), lty=c(1,2))
```
